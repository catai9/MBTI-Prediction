```python
import pandas as pd
from google.colab import drive
# Import data from file
drive.mount('/content/drive')
cnn = pd.read_csv('/content/drive/My Drive/MSCI 598 (Deep Learning for NLP) Project/Code/cnn_best_pred.csv')
cnn.head()
```

### Manually Selected Good Sentences

```python
# These will be manually selected by examining the text spoken by real MBTI categories.
cats = ["IE", "NS", "TF", "JP"]

def get_real_sentences(cnn):
  mbti = {}
  for cat in cats:
    i = 1
    for t in cat:
      mbti[t] = cnn[cnn[cat] == i][['text', 'mbti', "IE", "NS", "TF", "JP", 'IE_pred', 'NS_pred', 'TF_pred', 'JP_pred']]
      i -= 1
  return mbti

mbti_real = get_real_sentences(cnn)

t = "P"
sents = mbti_real[t]['text']
print("====================================================================================")
print("Type "+ t)
for s in sents:
  print(s)
```

### Sentences that have been categorized correctly

```python
# True Positives
cats = ["IE", "NS", "TF", "JP"]

def get_real_sentences(cnn):
  mbti = {}
  for cat in cats:
    i = 1
    for t in cat:
      mbti[t] = cnn[(cnn[cat] == i) & (cnn[cat+'_pred'] == i)][['text', 'mbti', "IE", "NS", "TF", "JP", 'IE_pred', 'NS_pred', 'TF_pred', 'JP_pred']]
      i -= 1
  return mbti

mbti_real = get_real_sentences(cnn)

t = "E"
sents = mbti_real[t]['text']
print("====================================================================================")
print("Type "+ t)
for s in sents:
  print(s)
```

### Sentences that have been predicted incorrectly

```python
# False Positives
cats = ["IE", "NS", "TF", "JP"]

def get_real_sentences(cnn):
  mbti = {}
  for cat in cats:
    i = 1
    for t in cat:
      mbti[t] = cnn[(cnn[cat] != i) & (cnn[cat+'_pred'] == i)][['text', 'mbti', "IE", "NS", "TF", "JP", 'IE_pred', 'NS_pred', 'TF_pred', 'JP_pred']]
      i -= 1
  return mbti

mbti_real = get_real_sentences(cnn)

t = "P"
sents = mbti_real[t]['text']
print("====================================================================================")
print("Type "+ t)
for s in sents:
  print(s)
```
